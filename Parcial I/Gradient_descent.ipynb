```python
import numpy as np
import matplotlib.pyplot as plt

#import sympy as sp #library -> numerical (symbols) calculus in python
#diff function -> spoiler

def f(theta):
    return (theta - 3) ** 2

def df(theta):
    return 2 * (theta - 3)  #differential function of (f-function)

#we need initialize the values:
#parameters (theta)
#learnig rate (eta)

theta , eta = 0.0 , 0.2963 #learning rate
thetas ,  values = [theta],[f(theta)] #list of stored values and f(theta)
print("Initial values:" ,theta, "function value (error)",f(theta))

#gradient descent
for t in range (20):
    print(f"Iteration {t}: theta = {theta:.6f}, error ={f(theta):.6f}")
    theta = theta - eta * df(theta) #theta update (based on eta (learning rate))
    thetas.append(theta) #list of parameters to graphs
    values.append(f(theta)) #list of values to graphs

#generate x values to draw the curve f(theta)
xspace = np.linspace(-10,15,500) #from -10 to 10 , 500 points

#plot the function f(theta)
plt.plot(xspace, f(xspace), label="f(theta-3)**2")
plt.plot(thetas, values, marker="o")#plot the path of gradinet with markers
plt.show()
```
