```python
import matplotlib.pyplot as plt  # to plot figures, images, graphs, etc.
from PIL import Image
import numpy as np

img_gray = np.array(Image.open("Images/lena256.png").convert("L"))
img_rgb = np.array(Image.open("Images/drone.png").convert("RGB"))  #nd-array
print(img_gray.shape)  #shpe 2 dimm
#print(img_gray)
print(img_rgb.shape)
np.savetxt("Datasets/imagen.csv", img_gray, delimiter=",", fmt="%.3f")
#np.savetxt("drone.csv", img_rgb, delimiter=",", fmt="%.3f") # dosen't work
#csv is 2 dimm file
#print(img_rgb)
```

```python
print("Grayscale (Lena)")
plt.figure()  #imagen space
plt.imshow(img_gray, cmap="gray")  #imagen show
plt.title("Grayscale Lena (L)")
plt.show()
```

```python
print("RGB Image (drone)")
plt.figure()  #imagen space
plt.imshow(img_rgb)  #imagen show
plt.title("RGB Image (drone)")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# rotate 90
rotate_img = np.rot90(img_rgb, 1)
#print(rotate_img)
print("RGB Rotate Image (drone)")
plt.figure()  #imagen space
plt.imshow(rotate_img)  #imagen show
plt.title("RGB Rotate Image (drone)")
plt.axis("off")
plt.show()

#horizontal flip (tensor)
#img_rgb is a n-dimensional array (tensor)
flip_horizontal = img_rgb[:, ::-1, :]
#img_rgb [] the first ":" -> all rows
#the second "::-1" take columns but in reverse order
#the third, ":" all RGB channels (R,G,B)
#this is like a right flip

print("Rotate Image (drone) horizontally")
plt.figure()
plt.imshow(flip_horizontal)  #imagen show
plt.title("Rotate Image (drone) horizontally")
plt.axis("off")
plt.show()

#vertical flip
flip_v = img_rgb[::-1, :, :]

print("Rotate Image (drone) vertically")
plt.figure()  #imagen space
plt.imshow(flip_v)  #imagen show
plt.title("Rotate Image (drone) vertically")
plt.axis("off")
plt.show()


#brighter
brighter =np.clip(flip_v+50,0,255).astype(np.uint8)
#clip forces all values so they stay between 0 and 255
#astype converts the result a 8-bit integers
print("Rotate Image (drone) vertically")
plt.figure()  #imagen space
plt.imshow(brighter)  #imagen show
plt.title("RGB Rotate Image (drone) Bright")
plt.axis("off")
plt.show()

#darker
darker = np.clip(flip_v-50,0,255).astype(np.uint8)
#clip forces all values so they stay between 0 and 255
#astype converts the result a 8-bit integers
print("Rotate Image (drone) vertically")
plt.figure()  #imagen space
plt.imshow(darker)  #imagen show
plt.title("RGB Rotate Image (drone) darker")
plt.axis("off")
plt.show()
```
